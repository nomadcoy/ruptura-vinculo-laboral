```python
#importação da base de dados
import pandas as pd
tabela = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")
```

### Construção do objeto empírico

Cada linha da base representa uma trajetória institucional condensada de um trabalhador.

Os dados expressam a forma como a organização observa, classifica e administra sua força de trabalho.

```python
# visualização da base
tabela.info()
```

A base revela quais dimensões do trabalho se tornam administrativamente visíveis e quais permanecem ausentes do registro institucional.

```python
# tratamento de dados
tabela = tabela.drop_duplicates()
tabela.isna().sum()
```

```python
# análise descritiva (ruptura)
tabela["Attrition"].value_counts()
tabela["Attrition"].value_counts(normalize=True)
```

```python
# Análise relacional (gráficos)

import plotly.express as px

colunas_analise = [
    "OverTime",
    "MonthlyIncome",
    "JobRole",
    "JobLevel",
    "YearsAtCompany",
    "WorkLifeBalance"
]

for coluna in colunas_analise:
    fig = px.histogram(
        tabela,
        x=coluna,
        color="Attrition",
        title=f"{coluna} x Ruptura do Vínculo Laboral",
        text_auto=True
    )
    fig.show()
```

```python
# Simulação institucional

tabela_simulada = tabela[
    (tabela["OverTime"] == "No") &
    (tabela["MonthlyIncome"] > tabela["MonthlyIncome"].median()) &
    (tabela["YearsAtCompany"] >= 3)
]

tabela_simulada["Attrition"].value_counts(normalize=True)
```

### Conclusão

A ruptura do vínculo laboral não emerge como falha individual,
mas como resposta racional a estruturas organizacionais desiguais.
